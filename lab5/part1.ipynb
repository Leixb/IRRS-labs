```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

np.random.seed(12342)


def set_theme():
    sns.set_theme(
        context="paper",
        style="whitegrid",
        font_scale=1.5,
        font="STIXGeneral",
        rc={
            "text.usetex": True,
        },
    )
    return {"dpi": 300}

save_args = set_theme()
```

```python
n = 500
m = 2
G_barabasi = nx.barabasi_albert_graph(n, m)

degree_freq = nx.degree_histogram(G_barabasi)
degrees = range(len(degree_freq))
#plt.figure(figsize=(12, 8)) 
plt.loglog(degrees[m:], degree_freq[m:],'go-') 
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()
```

```python

C = []
D = []
x = np.linspace(1, 500, 50)
for i in x:
    G = nx.erdos_renyi_graph(int(i), 0.5)
    #c = nx.average_clustering(G)

    try:
        d = nx.average_shortest_path_length(G)
    except nx.exception.NetworkXError:
        d = 0

    #C.append(c)
    D.append(d)

print(C)
print(D)

#C = np.array(C) / max(C)
D = np.array(D) / max(D)

#plt.plot(x, C, "r-")
#plt.savefig("clustering.png")
plt.plot(x, D, "b-")
plt.savefig("shortest_path.png")
```

```python
C = []
D = []

Csd = []
Dsd = []
x = np.linspace(0, 1, 50)
for i in x:
    td = []
    tc = []
    for j in range(10):
        G = nx.watts_strogatz_graph(500, 4, i)
        c = nx.average_clustering(G)
        try:
            d = nx.average_shortest_path_length(G)
        except nx.exception.NetworkXError:
            d = 0
            
        td.append(d)
        tc.append(c)
        print(i,d,c)

    C.append(np.mean(tc))
    D.append(np.mean(td))
    
    Csd.append(np.std(tc))
    Dsd.append(np.std(td))

C = np.array(C) / max(C)
D = np.array(D) / max(D)

Csd = np.array(Csd) / max(C)
Dsd = np.array(Dsd) / max(D)
```

```python
plt.plot(x, C, "r-", label="Clustering coefficient")
plt.plot(x, D, "b-", label="Avg. Shortest path")
plt.legend()
plt.xlabel("$p$")

plt.fill_between(x, C-np.sqrt(Csd), C+np.sqrt(Csd))
plt.fill_between(x, D-Dsd, D+Dsd)
plt.ylim((0,None))
#plt.savefig("figures/ws_model.pdf", **save_args)
```

```python
n, m = 5000, 4

G_barabasi = nx.barabasi_albert_graph(n, m)

degree_freq = nx.degree_histogram(G_barabasi)
degrees = range(len(degree_freq))
#plt.figure(figsize=(12, 8)) 
plt.loglog(degrees[m:], degree_freq[m:],'go') 
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.optimize import curve_fit

x = np.flatnonzero(degree_freq)
y = np.array(degree_freq)[x]
plt.loglog(x, y, 'go')

ylog = np.log(y)
```

```python
def func(x, a, g):
    return a*x**(-g)

popt, pcov = curve_fit(func, x, y)
```

```python
plt.plot(x, y, "o", label="Data")
plt.plot(x, func(x, *popt), 'r-',
         label='fit: a=%.0f, $\gamma$=%5.3f' % tuple(popt))

plt.xlabel("Degree ($k$)")
plt.ylabel("Frequency ($n\\cdot P(k)$)")
plt.ylim((1,None))
plt.loglog()
plt.legend()
plt.savefig("figures/ba_fit.pdf")
```
